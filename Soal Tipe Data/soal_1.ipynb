```python
import numpy as np

# Definisi struktur data
company_detail_list = {
    'Company 1': {
        'name': 'Company 1',
        'domain': 'Retail',
        'clients': [
            {
                'name': 'acme.inc',
                'country': 'united states'
            },
            {
                'name': 'Wayne.co',
                'country': 'united states'
            }
        ]
    },
    'Company 2': {
        'name': 'Company 2',
        'domain': 'Construction',
        'clients': [
            {
                'name': 'Tesla',
                'country': 'united states'
            },
            {
                'name': 'Japan Airlines',
                'country': 'japan'
            },
            {
                'name': 'Indofood',
                'country': 'indonesia'
            }
        ]
    },
    'Company 3': {
        'name': 'Company 3',
        'domain': 'Healthcare',
        'clients': [
            {
                'name': 'Petronas',
                'country': 'malaysia'
            },
            {
                'name': 'VW Group',
                'country': 'germany'
            },
            {
                'name': 'IBM',
                'country': 'united states'
            },
            {
                'name': 'Mitsubishi',
                'country': 'japan'
            }
        ]
    }
}

employee_detail_list = {
    'John Doe': {
        'name': 'EMP-0001',
        'first_name': 'John',
        'last_name': 'Doe',
        'full_name': 'John Doe',
        'company': 'Company 1'
    },
    'Tom Smith': {
        'name': 'EMP-0002',
        'first_name': 'Tom',
        'last_name': 'Smith',
        'full_name': 'Tom Smith',
        'company': 'Company 2'
    },
    'Andrew Sebastian': {
        'name': 'EMP-0003',
        'first_name': 'Andrew',
        'last_name': 'Sebastian',
        'full_name': 'Andrew Sebastian',
        'company': 'Company 2'
    },
}

# Mencari field-field yang memiliki data kosong dalam company_detail_list
empty_fields_company = {}
for company_name, company_details in company_detail_list.items():
    for key, value in company_details.items():
        if isinstance(value, list):
            for item in value:
                if any(isinstance(v, dict) and 'name' in v and 'country' in v for v in item.values()):
                    for client in item:
                        if 'name' not in client or 'country' not in client:
                            if key not in empty_fields_company:
                                empty_fields_company[key] = []
                            empty_fields_company[key].append((company_name, client))

# Menghitung nilai rata-rata untuk setiap field yang memiliki data kosong
for field, empty_list in empty_fields_company.items():
    total = 0
    count = 0
    for company_name, client in empty_list:
        total += len(company_detail_list[company_name][field])
        count += 1
    mean = total / count

    # Mengganti data kosong dengan nilai rata-rata
    for company_name, client in empty_list:
        company_detail_list[company_name][field][client] = mean

# Mencari field-field yang memiliki data kosong dalam employee_detail_list
empty_fields_employee = {}
for employee_name, employee_details in employee_detail_list.items():
    for key, value in employee_details.items():
        if value is None:
            if key not in empty_fields_employee:
                empty_fields_employee[key] = []
            empty_fields_employee[key].append(employee_name)

# Menghitung nilai rata-rata untuk setiap field yang memiliki data kosong
for field, empty_list in empty_fields_employee.items():
    total = 0
    count = 0
    for employee_name in empty_list:
        total += len(employee_detail_list[employee_name][field])
        count += 1
    mean = total / count

    # Mengganti data kosong dengan nilai rata-rata
    for employee_name in empty_list:
        employee_detail_list[employee_name][field] = mean

# Mencetak nilai rata-rata untuk setiap field yang diisi
print("Nilai rata-rata untuk setiap field yang diisi:")
for field, mean in empty_fields_company.items():
    print(f"{field}: {mean}")

for field, mean in empty_fields_employee.items():
    print(f"{field}: {mean}")

print("Data kosong telah diisi dengan nilai rata-rata.")
```

```python
import pandas as pd

# Definisi struktur data perusahaan
company_detail_list = {
    'Company 1': {
        'name': 'Company 1',
        'domain': 'Retail',
        'clients': [
            {
                'name': 'acme.inc',
                'country': 'united states'
            },
            {
                'name': 'Wayne.co',
                'country': 'united states'
            }
        ]
    },
    'Company 2': {
        'name': 'Company 2',
        'domain': 'Construction',
        'clients': [
            {
                'name': 'Tesla',
                'country': 'united states'
            },
            {
                'name': 'Japan Airlines',
                'country': 'japan'
            },
            {
                'name': 'Indofood',
                'country': 'indonesia'
            }
        ]
    },
    'Company 3': {
        'name': 'Company 3',
        'domain': 'Healthcare',
        'clients': [
            {
                'name': 'Petronas',
                'country': 'malaysia'
            },
            {
                'name': 'VW Group',
                'country': 'germany'
            },
            {
                'name': 'IBM',
                'country': 'united states'
            },
            {
                'name': 'Mitsubishi',
                'country': 'japan'
            }
        ]
    }
}

# Fungsi untuk mendapatkan karyawan beserta domain perusahaan mereka
def get_employees(employee_detail_list, company_detail_list):
    employees = []
    for employee, details in employee_detail_list.items():
        company_name = details['company']
        domain = company_detail_list.get(company_name, {}).get('domain', 'Unknown')
        employee_data = {
            'Employee': employee,
            'Domain': domain
        }
        employees.append(employee_data)
    return employees

# Memanggil fungsi get_employees
employees_data = get_employees(employee_detail_list, company_detail_list)

# Membuat DataFrame dari hasil
employees_df = pd.DataFrame(employees_data)

# Mencetak DataFrame untuk menampilkan karyawan beserta domain perusahaannya
print("\nKaryawan dan Domain Perusahaan:")
print(employees_df)

```

```python
import pandas as pd

# Definisi struktur data perusahaan
company_detail_list = {
    'Company 1': {
        'name': 'Company 1',
        'domain': 'Retail',
        'clients': [
            {
                'name': 'acme.inc',
                'country': 'united states'
            },
            {
                'name': 'Wayne.co',
                'country': 'united states'
            }
        ]
    },
    'Company 2': {
        'name': 'Company 2',
        'domain': 'Construction',
        'clients': [
            {
                'name': 'Tesla',
                'country': 'united states'
            },
            {
                'name': 'Japan Airlines',
                'country': 'japan'
            },
            {
                'name': 'Indofood',
                'country': 'indonesia'
            }
        ]
    },
    'Company 3': {
        'name': 'Company 3',
        'domain': 'Healthcare',
        'clients': [
            {
                'name': 'Petronas',
                'country': 'malaysia'
            },
            {
                'name': 'VW Group',
                'country': 'germany'
            },
            {
                'name': 'IBM',
                'country': 'united states'
            },
            {
                'name': 'Mitsubishi',
                'country': 'japan'
            }
        ]
    }
}

# Fungsi untuk mendapatkan daftar karyawan yang bekerja di setiap perusahaan
def get_employees_by_company(employee_detail_list):
    employees_by_company = {}
    for employee, details in employee_detail_list.items():
        company_name = details['company']
        if company_name in employees_by_company:
            employees_by_company[company_name].append(employee)
        else:
            employees_by_company[company_name] = [employee]
    return employees_by_company

# Memanggil fungsi get_employees_by_company
employees_by_company_data = get_employees_by_company(employee_detail_list)

# Membuat DataFrame dari hasil
employees_by_company_df = pd.DataFrame.from_dict(employees_by_company_data, orient='index')

# Mencetak DataFrame untuk menampilkan daftar karyawan per perusahaan
print("\nDaftar Karyawan per Perusahaan:")
print(employees_by_company_df)

```
