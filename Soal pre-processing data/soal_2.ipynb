```python
import pandas as pd

# Baca file CSV ke dalam DataFrame
data_startup = pd.read_csv('50_Startups.csv')

# Isi data yang kosong dengan nilai rata-rata dari kolom 'R&D Spend', 'Administration', 'Marketing Spend', dan 'Profit'
data_startup['R&D Spend'].fillna(round(data_startup['R&D Spend'].mean(), 2), inplace=True)
data_startup['Administration'].fillna(round(data_startup['Administration'].mean(), 2), inplace=True)
data_startup['Marketing Spend'].fillna(round(data_startup['Marketing Spend'].mean(), 2), inplace=True)
data_startup['Profit'].fillna(round(data_startup['Profit'].mean(), 2), inplace=True)

# Tampilkan DataFrame setelah pengisian nilai kosong
print("\nDataFrame setelah pengisian data kosong:")
data_startup.sample(7)
```

```python
import pandas as pd

# Baca file CSV ke dalam DataFrame
data_startup = pd.read_csv('50_Startups.csv')

# Isi data yang kosong dengan nilai rata-rata dari kolom 'R&D Spend', 'Administration', 'Marketing Spend', dan 'Profit'
data_startup['R&D Spend'].fillna(round(data_startup['R&D Spend'].mean(), 2), inplace=True)
data_startup['Administration'].fillna(round(data_startup['Administration'].mean(), 2), inplace=True)
data_startup['Marketing Spend'].fillna(round(data_startup['Marketing Spend'].mean(), 2), inplace=True)
data_startup['Profit'].fillna(round(data_startup['Profit'].mean(), 2), inplace=True)

# Lakukan One-Hot Encoding pada kolom 'State'
data_startup_encoded = pd.get_dummies(data_startup, columns=['State'])

# Tampilkan 7 baris pertama dari DataFrame setelah pengisian nilai kosong dan One-Hot Encoding
print("\nBaris pertama DataFrame setelah pengisian data kosong dan One-Hot Encoding: \n")
print(data_startup_encoded.head(7))
```

```python
import pandas as pd

# Baca file CSV ke dalam DataFrame
data_startup = pd.read_csv('50_Startups.csv')

# Isi data yang kosong dengan nilai rata-rata dari kolom 'R&D Spend', 'Administration', 'Marketing Spend', dan 'Profit'
data_startup['R&D Spend'].fillna(round(data_startup['R&D Spend'].mean(), 2), inplace=True)
data_startup['Administration'].fillna(round(data_startup['Administration'].mean(), 2), inplace=True)
data_startup['Marketing Spend'].fillna(round(data_startup['Marketing Spend'].mean(), 2), inplace=True)
data_startup['Profit'].fillna(round(data_startup['Profit'].mean(), 2), inplace=True)

# Buat kolom baru 'Tax' berdasarkan rumus yang diberikan
data_startup = data_startup.assign(Tax=lambda x: (x['Profit'] + x['Marketing Spend'] + x['Administration']) * 0.05)

# Tampilkan 7 baris pertama DataFrame setelah penambahan kolom 'Tax'
print("\nBaris pertama DataFrame setelah penambahan kolom 'Tax':")
print(data_startup.head(7))
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Baca file CSV ke dalam DataFrame
data_startup = pd.read_csv('50_Startups.csv')

# Isi data yang kosong dengan nilai rata-rata dari kolom 'R&D Spend', 'Administration', 'Marketing Spend', dan 'Profit'
data_startup['R&D Spend'].fillna(round(data_startup['R&D Spend'].mean(), 2), inplace=True)
data_startup['Administration'].fillna(round(data_startup['Administration'].mean(), 2), inplace=True)
data_startup['Marketing Spend'].fillna(round(data_startup['Marketing Spend'].mean(), 2), inplace=True)
data_startup['Profit'].fillna(round(data_startup['Profit'].mean(), 2), inplace=True)

# Buat objek StandardScaler
scaler = StandardScaler()

# Lakukan fit_transform pada DataFrame untuk menghitung mean dan standar deviasi dari setiap fitur
# dan kemudian melakukan transformasi data
data_scaled = scaler.fit_transform(data_startup[['R&D Spend', 'Administration', 'Marketing Spend', 'Profit']])

# Simpan hasil transformasi ke dalam DataFrame baru
data_scaled_df = pd.DataFrame(data_scaled, columns=['R&D Spend', 'Administration', 'Marketing Spend', 'Profit'])

# Buat kolom 'Tax' berdasarkan rumus yang diberikan
data_scaled_df['Tax'] = (data_scaled_df['Profit'] + data_scaled_df['Marketing Spend'] + data_scaled_df['Administration']) * 0.05

# Tampilkan 7 baris pertama DataFrame setelah transformasi
print("\n7 baris pertama DataFrame setelah StandardScaler:")
print(data_scaled_df.head(7))
```
